```python
import pdfplumber


pdf=pdfplumber.open("EDU.pdf")
page = pdf.pages[0]
table=page.extract_table({
"vertical_strategy": "lines_strict", 
"horizontal_strategy": "lines_strict",})

img=page.to_image()    
gridlines =  [ x for x in page.lines
        if x["width"] > 10 and x["width"] < 200 ]
img.reset().draw_lines(gridlines)
display(img)
```

```python
crop = page.within_bbox((0, 110, page.width, page.height-60))
img=crop.to_image()
#page.extract_words()
tablelines=[]
for i in range(len(gridlines)):
    if i >0 and gridlines[i]['y0']==gridlines[i-1]['y0']:
        continue
    #img.reset().draw_line(gridlines[i])
    #display(img)
    tablelines.append(gridlines[i])
img.reset().draw_lines(tablelines)
display(img)
```
